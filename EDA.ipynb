```python
import pandas as pd
import numpy as np
import random

np.random.seed(42)
n = 50

names = [
    'Aarav', 'Priya', 'Rohit', 'Sneha', 'Arjun', 'Pooja', 'Vikram', 'Neha',
    'Karan', 'Anjali', 'Rahul', 'Divya', 'Amit', 'Shreya', 'Suresh', 'Kavya',
    'Ravi', 'Meera', 'Aditya', 'Nisha', 'Varun', 'Simran', 'Manish', 'Ritika',
    'Deepak', 'Tanvi', 'Sanjay', 'Komal', 'Nikhil', 'Ananya'
]

df = pd.DataFrame({
    'name': [random.choice(names) for _ in range(n)],
    'age': np.random.randint(21, 65, n),
    'gender': np.random.choice(['Male', 'Female'], n),
    'income': np.random.randint(20000, 200000, n),
    'loan_amount': np.random.randint(5000, 100000, n),
    'credit_score': np.random.randint(300, 850, n),
    'employment_years': np.random.randint(0, 30, n),
    'education': np.random.choice(['Graduate', 'Not Graduate'], n),
    'property_area': np.random.choice(['Urban', 'Rural', 'Semiurban'], n),
    'married': np.random.choice(['Yes', 'No'], n),
    'loan_status': np.random.choice(['Approved', 'Rejected', 'Pending'], n, p=[0.6, 0.3, 0.1])
})

df.loc[np.random.choice(n, 50), 'credit_score'] = np.nan
df.loc[np.random.choice(n, 30), 'income'] = np.nan
df.loc[np.random.choice(n, 20), 'employment_years'] = np.nan
df.index = range(1, n+1)
print(df.shape)
df.head(50)
```

```python
#basics of EDA using pandas
   
print("shape :\n",df.shape)
print("top 5 rows are :\n",df.head())
print("bottom 5 rows are :\n",df.tail())
print(df.info())  
print(df.describe()) 
print(df.dtypes )
```

```python
print( df.duplicated().sum())
print(df.isnull())
print(df.isnull().sum())
```

```python
print(df['credit_score'].isnull().sum())
```

```python
median=df['credit_score'].median()
print(median)
mean=df['income'].mean()
print(mean)
med=df['employment_years'].median()
print(med)
```

```python
df['credit_score'] = df['credit_score'].fillna(df['credit_score'].median())
df['income'] = df['income'].fillna(df['income'].median())
df['employment_years'] = df['employment_years'].fillna(df['employment_years'].median())


print(df.isnull().sum())
```
